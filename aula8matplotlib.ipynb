```python
#EXEMPLO USANDO UM DATASET
from sklearn.datasets import load_iris
import pandas as pd
data = load_iris()
df= pd.DataFrame(data.data, columns=data.feature_names)
df['target'] = data.target
plt.scatter(df['petal length (cm)'],df['petal width (cm)'],c=df['target'])
print(df)
```
