```python
from sklearn.datasets import load_iris
import pandas as pd
```

```python
data = load_iris()
df = pd.DataFrame(data.data, columns=data.feature_names)
df['target'] = data.target
df
```

```python
from matplotlib import pyplot as plt
plt.scatter(df['sepal length (cm)'], df['sepal width (cm)'], c=df['target'])
plt.show()
```

```python
# Crie um histograma de uma lista de idade dos alunos da sala
idades = [17, 18, 18, 18, 18, 18, 19, 20, 19, 20, 21, 23, 5, 19, 17, 5, 40, 24, 22, 20]

# Exiba o histograma
plt.hist(idades)
plt.show()
```

```python
# Crie um histograma com 5 colunas da lista de idades
plt.hist(idades, bins=5)

# Exiba o histograma e limpe o grfico
plt.show()
plt.clf()


# Faa um novo histograma com 20 colunas
plt.hist(idades, bins=20)

# Exiba o histograma e limpe o grfico
plt.show()
plt.clf
```
