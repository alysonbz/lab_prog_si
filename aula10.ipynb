customização de gráficos

```python
import matplotlib.pyplot as plt
year = [1950,1970,1990,2010]
pop = [2.519,3.692,5.263,6.972]
plt.plot(year, pop)
plt.xlabel("ano")
plt.ylabel("população")
plt.title("projeção da populção do mundo")
plt.yticks([0,2,4,6,8,10],
           ["0","2B","4B","6B","8B","10B"])
plt.show()
```
